```python
CONTESTS = set(contests_df["contest_name"])
TEAMS = set(contests_df["team_name"])
WORKERS = set(contests_df["n_workers"])
```

```python
for contest in CONTESTS:
    for team in TEAMS:
        data = {s['n_workers']: s['avarage'] for _, s in contests_df.iterrows() 
             if s['contest_name'] == contest and s['team_name'] == team}
        plt.title(contest)
        plt.plot(data.keys(), data.values(), '-o', label=team)
        plt.xlabel('num of workers')
        plt.ylabel('avarage time(s)')
        plt.legend(loc='best')
    plt.show()
```
